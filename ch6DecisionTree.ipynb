```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
tree_clf.predict_proba([[5, 1.5]])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_coin_tosses(num_tosses=1000):
    # 0 = Head, 1 = Tail
    tosses = np.random.randint(0, 2, size=num_tosses)
    
    # Count cumulative heads and tails
    heads = np.cumsum(tosses == 0)
    tails = np.cumsum(tosses == 1)
    
    # Probabilities after each toss
    total = np.arange(1, num_tosses + 1)
    prob_heads = heads / total
    prob_tails = tails / total

    # Final count
    final_heads = heads[-1]
    final_tails = tails[-1]
    print(f"Total Tosses: {num_tosses}")
    print(f"Heads: {final_heads}, Tails: {final_tails}")
    print(f"Head Probability: {prob_heads[-1]:.4f}, Tail Probability: {prob_tails[-1]:.4f}")

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(total, prob_heads, label='Probability of Heads', color='blue')
    plt.plot(total, prob_tails, label='Probability of Tails', color='green')
    plt.axhline(y=0.5, color='red', linestyle='--', label='Ideal Probability = 0.5')
    plt.xlabel("Number of Tosses")
    plt.ylabel("Probability")
    plt.title("Coin Toss Probabilities Over Time")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Run the simulation with 5000 tosses
simulate_coin_tosses(num_tosses=1000000)
```
